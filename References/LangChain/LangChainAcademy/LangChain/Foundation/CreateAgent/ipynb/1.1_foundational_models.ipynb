## Initialising and invoking a model

```python
import os
import time
import pprint
from langchain import chat_models

# Initialize using the OpenAI provider with a custom base_url
llm = chat_models.init_chat_model(
    model="gpt-oss:120b-cloud",
    model_provider="openai",
    base_url="https://ollama.com/v1",
    api_key=os.environ["OLLAMA_API_KEY"],
    temperature=0
)

start_time = time.time()
response = llm.invoke(input="""
    What's the capital of the Moon?
""")
end_time = time.time() - start_time
print(f"Time taken: %.2fs seconds"%(end_time))

print(response.content)
pprint.pprint(response.response_metadata)
```

## Initialising and invoking an agent

```python
import os
import time
from langchain import chat_models, agents, messages

# Initialize using the OpenAI provider with a custom base_url
llm = chat_models.init_chat_model(
    model="gpt-oss:120b-cloud",
    model_provider="openai",
    base_url="https://ollama.com/v1",
    api_key=os.environ["OLLAMA_API_KEY"],
    temperature=0
)

agent = agents.create_agent(model=llm)

start_time = time.time()
response = agent.invoke(
    input={"messages": [
        messages.HumanMessage(content="""
            What's the capital of the Moon?
        """)
    ]}
)
end_time = time.time() - start_time
print(f"Time taken: %.2fs seconds"%(end_time))

for m in response['messages']:
    m.pretty_print()
```

```python
import os
import time
from langchain import chat_models, agents, messages

# Initialize using the OpenAI provider with a custom base_url
llm = chat_models.init_chat_model(
    model="gpt-oss:120b-cloud",
    model_provider="openai",
    base_url="https://ollama.com/v1",
    api_key=os.environ["OLLAMA_API_KEY"],
    temperature=0
)

agent = agents.create_agent(model=llm)

start_time = time.time()
response = agent.invoke(
    input={"messages": [
        messages.HumanMessage(content="""
            What's the capital of the Moon?
        """),
        messages.AIMessage(content="""
            The capital of the Moon is Luna City.
        """),
        messages.HumanMessage(content="""
            Interesting, tell me more about Luna City.
        """)
    ]}
)
end_time = time.time() - start_time
print(f"Time taken: %.2fs seconds"%(end_time))

for m in response['messages']:
    m.pretty_print()
```

## Streaming Output

```python
import os
import time
from langchain import chat_models, agents, messages

# Initialize using the OpenAI provider with a custom base_url
llm = chat_models.init_chat_model(
    model="gpt-oss:120b-cloud",
    model_provider="openai",
    base_url="https://ollama.com/v1",
    api_key=os.environ["OLLAMA_API_KEY"],
    temperature=0
)

agent = agents.create_agent(model=llm)

start_time = time.time()
for token, metadata in agent.stream(
    input={"messages": [
        messages.HumanMessage(content="""
            Tell me all about Luna City, the capital of the Moon.
        """)
    ]},
    stream_mode="messages"
):
    # token is a message chunk with token content
    # metadata contains which node produced the token
    if token.content:  # Check if there's actual content
        print(token.content, end="", flush=True)  # Print token

end_time = time.time() - start_time
print(f"Time taken: %.2fs seconds"%(end_time))
```
