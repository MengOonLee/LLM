```python
import warnings
warnings.filterwarnings("ignore")
import dotenv

_ = dotenv.load_dotenv(dotenv_path=".env", override=True)
```

```python
import os
import time
from langchain import chat_models, agents, messages
from dataclasses import dataclass

llm = chat_models.init_chat_model(
    model="gpt-oss:120b-cloud",
    model_provider="openai",
    base_url="https://ollama.com/v1",
    api_key=os.environ["OLLAMA_API_KEY"],
    temperature=0
)

@dataclass
class ColourContext:
    favourite_colour: str = "blue"
    least_favourite_colour: str = "yellow"

agent = agents.create_agent(
    model=llm,
    context_schema=ColourContext
)

start_time = time.time()
response = agent.invoke(
    input={"messages": [messages.HumanMessage(content="""
        What is my favourite colour?
    """)]},
    context=ColourContext()
)
end_time = time.time() - start_time
print(f"Time taken: %.2fs"%(end_time))

for m in response["messages"]:
    m.pretty_print()
```

## Accessing Context

```python
import os
import time
from langchain import chat_models, tools, agents, messages
from dataclasses import dataclass

llm = chat_models.init_chat_model(
    model="gpt-oss:120b-cloud",
    model_provider="openai",
    base_url="https://ollama.com/v1",
    api_key=os.environ["OLLAMA_API_KEY"],
    temperature=0
)

@dataclass
class ColourContext:
    favourite_colour: str = "blue"
    least_favourite_colour: str = "yellow"

@tools.tool
def get_favourite_colour(runtime: tools.ToolRuntime) -> str:
    """
    Get the favourite colour of the user
    """
    return runtime.context.favourite_colour

@tools.tool
def get_least_favourite_colour(runtime: tools.ToolRuntime) -> str:
    """
    Get the least favourite colour of the user
    """
    return runtime.context.least_favourite_colour

agent = agents.create_agent(
    model=llm,
    tools=[get_favourite_colour, get_least_favourite_colour],
    context_schema=ColourContext
)

start_time = time.time()
response = agent.invoke(
    input={"messages": [messages.HumanMessage(content="""
        What is my favourite colour?
    """)]},
    context=ColourContext()
)
end_time = time.time() - start_time
print(f"Time taken: %.2fs"%(end_time))

for m in response["messages"]:
    m.pretty_print()
```

```python
import os
import time
from langchain import chat_models, tools, agents, messages
from dataclasses import dataclass

llm = chat_models.init_chat_model(
    model="gpt-oss:120b-cloud",
    model_provider="openai",
    base_url="https://ollama.com/v1",
    api_key=os.environ["OLLAMA_API_KEY"],
    temperature=0
)

@dataclass
class ColourContext:
    favourite_colour: str = "blue"
    least_favourite_colour: str = "yellow"

@tools.tool
def get_favourite_colour(runtime: tools.ToolRuntime) -> str:
    """
    Get the favourite colour of the user
    """
    return runtime.context.favourite_colour

@tools.tool
def get_least_favourite_colour(runtime: tools.ToolRuntime) -> str:
    """
    Get the least favourite colour of the user
    """
    return runtime.context.least_favourite_colour

agent = agents.create_agent(
    model=llm,
    tools=[get_favourite_colour, get_least_favourite_colour],
    context_schema=ColourContext
)

start_time = time.time()
response = agent.invoke(
    input={"messages": [messages.HumanMessage(content="""
        What is my least favourite colour?
    """)]},
    context=ColourContext(least_favourite_colour="black")
)
end_time = time.time() - start_time
print(f"Time taken: %.2fs"%(end_time))

for m in response["messages"]:
    m.pretty_print()
```
